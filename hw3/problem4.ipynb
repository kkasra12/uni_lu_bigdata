```python
from typing import Callable


def make_results(n_splits:int, x_data: pd.DataFrame, y_data: pd.DataFrame, base_model: Callable, metrics: list):

    raw_results = {
        mtric.__name__.rsplit("_", 1)[0]: [] for mtric in metrics
    }
    for train_index, test_index in KFold(n_splits=n_splits).split(befor_preprocess):
        x_train, x_test = x_data.iloc[train_index], x_data.iloc[test_index]
        y_train, y_test = y_data.iloc[train_index], y_data.iloc[test_index]
        model = base_model()
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        for metric in metrics:
            try:
                m = metric(y_test, y_pred, average="weighted")
            except TypeError:
                m = metric(y_test, y_pred)
            raw_results[metric.__name__.rsplit("_", 1)[0]].append(m)
    final_results = {}
    for metric in metrics:
        metric_name = metric.__name__.rsplit("_", 1)[0]
        final_results[metric_name] = {
            "min": min(raw_results[metric_name]),
            "max": max(raw_results[metric_name]),
            "mean": np.mean(raw_results[metric_name]),
            "stddev": np.std(raw_results[metric_name]),
        }
    return final_results
```

```python
def accuracy_score_(y_true, y_pred, average):
    return accuracy_score(y_true, y_pred)
```

```python
regression_metrics = [r2_score, mean_squared_error, mean_absolute_error]
classification_metrics = [f1_score, precision_score, recall_score, accuracy_score]
```

```python
x_data = befor_preprocess.drop(columns=["S1Q061"])
y_data = befor_preprocess["S1Q061"]
dataset_b_results_s1q061 = make_results(n_splits=30,
                                 x_data=x_data,
                                 y_data=y_data,
                                 base_model=DecisionTreeClassifier,
                                 metrics=classification_metrics)
dataset_b_results_s1q061
```
